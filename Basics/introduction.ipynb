# Matplotlib Introduction

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [20, 40, 50, 80, 100]

plt.plot(x, y)
plt.grid(True)
plt.show()
```

```python
x = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
y = [24, 32, 39, 36, 40]

plt.plot(x, y)
plt.title('Sales Data')

plt.xlabel('Day of the Week')
plt.ylabel('Sales per day')

plt.show()
```

```python
months = [1, 2, 3, 4, 5, 6]
sales = [2000, 4000, 3000, 8400, 7100, 6560]

plt.plot(months, sales, color='green', linestyle='--', linewidth=2, marker='o', label='2025 Sales Data')
plt.title('Sales Data')

plt.xlabel('Month')
plt.ylabel('Sales')

plt.grid(color='blue', linestyle=':', linewidth=0.5)
plt.legend(loc='upper left', fontsize=20)

plt.xlim(1, 4)
plt.ylim(0, 5000)

plt.xticks([1, 2, 3, 4], ['Jan', 'Feb', 'Mar', 'Apr'])
plt.show()
```

```python
product = ['A', 'B', 'C', 'D', 'E']
sales = [2000, 4000, 3000, 8400, 7100]

plt.bar(product, sales, color='orange', label='2025 Sales Data')
# to make this howrizontal -> plt.barh()
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sales Data Comparison')
plt.legend()
plt.show()
```

```python
regions = ['North', 'South', 'East', 'West']
revenue = [100000, 120000, 80000, 150000]

plt.pie(revenue, labels=regions, autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'lightcoral', 'gold'])
plt.title('Revenue by Region')
plt.show()
```

```python
scores = [85, 90, 58, 62, 88, 76, 95, 69, 82, 87, 50, 79, 51, 74, 86, 93, 81, 77, 94, 63]

plt.hist(scores, bins=5, color='skyblue', edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel('Number of Students')
plt.title('Student Scores Distribution')
plt.show()
```

```python
hours_studied = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores = [60, 65, 70, 75, 80, 85, 90, 95]

plt.scatter(hours_studied, exam_scores, color='red', marker='o', label='Exam Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.title('Relationship between Hours Studied and Exam Scores')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.scatter([1, 2, 3], [50, 55, 60], color='blue', label='Group A')
plt.scatter([1, 2, 3], [45, 50, 52], color='green', label='Group B')

plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Comparison of Exam Scores between Two Groups')
plt.legend()
plt.grid()
plt.show()
```
