# Matplotlib Introduction

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [20, 40, 50, 80, 100]

plt.plot(x, y)
plt.grid(True)
plt.show()
```

```python
x = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
y = [24, 32, 39, 36, 40]

plt.plot(x, y)
plt.title('Sales Data')

plt.xlabel('Day of the Week')
plt.ylabel('Sales per day')

plt.show()
```
